```
import numpy as np
import matplotlib.pyplot as plt
```

```
def f(w,b,x):
	return 1.0/(1.0+np.exp(-(w*x+b)))
```

```
x = np.arange(-6.0, 6.0, 0.2)

y = f(5,1,x)
plt.plot(x,y)
plt.show()
# for w increase
```

```
x = np.arange(-6.0, 6.0, 0.2)

y = f(1,1,x)
plt.plot(x,y)
plt.show()
```

```
# for b increase
x = np.arange(-6.0, 6.0, 0.2)

y = f(1,6,x)
plt.plot(x,y)
plt.show()
```

```
# 6. plot function f21

def f11(a,b):
  return 1.0/(1.0+np.exp(-((400*a) + b)))

def f12(a,b):
  return 1.0/(1.0+np.exp(-((400*a) - b)))

def f21(a,b):
  return f11(a,b) - f12(a,b)

b = 24

# x[-1:1]
x = np.arange(-1.0, 1.0, 0.2)

plt.plot(x,f21(x,b))
plt.show()
# answer graph 3
```

```
# 7th 
def h11(x1,x2):
  return 1.0/(1.0+np.exp(-(x1 + (100*x2) + 200)))

def h12(x1,x2):
  return 1.0/(1.0+np.exp(-(x1 + (100*x2) - 200)))

def h13(x1,x2):
  return 1.0/(1.0+np.exp(-( (100*x1) + (x2) + 200)))

def h14(x1,x2):
  return 1.0/(1.0+np.exp(-( (x1*100) + x2 - 200)))

def h21(x1,x2):
  return h11(x1,x2) - h12(x1,x2)

def h22(x1,x2):
  return h13(x1,x2) - h14(x1,x2)

def h31(x1,x2):
  return h21(x1,x2) - h22(x1,x2)

def f(x1,x2):
  return 1.0/(1.0+np.exp(-((50*h31(x1,x1)) - 100)))
```
